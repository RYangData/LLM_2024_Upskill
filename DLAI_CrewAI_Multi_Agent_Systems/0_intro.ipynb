## 4. Key Elements of AI Agents



# 1. RolePlaying

- Giving it a role/title and give it a shape/form/function for a task.. it changes system behaviour by a lot! 



# 2. Focus
- Currently we have large context length. and a lot of distractions
- If there are too many tasks, we know LLMs will not be good at all of them
- Agents.. 
    - need to focus on one thing or a few things! 
    
    


# 3. Tools
- Dont provide agents with EVERY tool
- Only provide them with tools required to do the job right! 



# 4. Co-operation
- Talk to each other and simulate actual feedback and delegation of tasks.
- LLMs are surprisingly quite competent at reflection
- We need to make sure agents are set-up so they can collaborate with each other. 

# 5 Guardrails


- Agents were stuck trying to use the same tool
- These are important and are easily integrated in CrewAI

# 6 Memory
- ABility for agent to remember what it has done
- What has it done in the past? 
- Learn from it and improve/dont make the same mistake again.
- Short term meory - can tap into learning from other agents during exeuction
- Long term memoery - after execution finishes... itself critiques itself ..
